```python
#Importando as Bibliotecas
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#carregamento da imagem
imagem_exemplo= cv2.imread('MRI_images/Dataset/Mild_Demented/mild_100.jpg')
```

```python
# FILTRO DE MÉDIA

def media_filtro(imagem, kernel_size=3):
    return cv2.blur(imagem, (kernel_size, kernel_size))
```

```python
#aplicação do filtro 
imagem_filtrada = media_filtro(imagem_exemplo,kernel_size=3)
```

```python
# Exibir a imagem antes e depois
plt.subplot(1, 2, 1), plt.imshow(imagem_exemplo, cmap='gray')
plt.title('Imagem Original'), plt.xticks([]), plt.yticks([])

plt.subplot(1, 2, 2), plt.imshow(imagem_filtrada, cmap='gray')
plt.title('Imagem com Filtro de Média'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
  # Aplicando o Filtro Sobel
sobelx = cv2.Sobel(imagem_exemplo, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(imagem_exemplo, cv2.CV_64F, 0, 1, ksize=5)
```

```python
# Exibir a imagem antes e depois do Filtro Sobel
plt.subplot(1, 3, 1), plt.imshow(imagem_exemplo, cmap='gray')
plt.title('Imagem Original'), plt.xticks([]), plt.yticks([])

plt.subplot(1, 3, 2), plt.imshow(sobelx, cmap='gray')
plt.title('Imagem Sobel em X'), plt.xticks([]), plt.yticks([])


plt.subplot(1, 3, 3), plt.imshow(sobely, cmap='gray')
plt.title('Imagem Sobel em Y'), plt.xticks([]), plt.yticks([])

plt.show()
```
